```python
import torch
import numpy as np
from scipy.ndimage.filters import median_filter, gaussian_filter
from scipy.signal import convolve2d, correlate2d
from skimage.util import random_noise
import os
import shutil
```

```python
class Example:
    def __init__(self, full_identifier, label):
        self.full_identifier = full_identifier
        self.label = label

    def __repr__(self):
        return (f"Example(full_identifier='{self.full_identifier}', label={self.label})")

def parse_file(file_path):
    objects = []
    with open(file_path, 'r') as file:
        for line in file:
            # Split the line
            parts = line.strip().split()
            full_identifier = parts[0]  # Keep the full identifier (e.g., 'american_bulldog_125')
            label = list(map(int, parts[1:]))  # Convert the label parts into a list of integers
            objects.append(Example(full_identifier, label))
    return objects


# Example usage
trainval_txt_path = r'../annotations/trainval.txt'
test_txt_path = r'../annotations/test.txt'

trainval_txt = parse_file(trainval_txt_path)
test_txt = parse_file(test_txt_path)

print(trainval_txt[100])
```

```python
def copy_matching_pictures(examples, source_folder, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)  # Create the destination folder if it doesn't exist

    for example in examples:
        picture_name = f"{example.full_identifier}.jpg"  # Construct the expected picture file name
        source_path = os.path.join(source_folder, picture_name)
        destination_path = os.path.join(destination_folder, picture_name)
        
        if os.path.exists(source_path):
            shutil.copy(source_path, destination_path)  # Copy the file to the destination folder
            print(f"Copied: {picture_name} to {destination_folder}")
        else:
            print(f"File not found: {source_path}")

source_folder = '../images'  # Replace with the folder containing pictures

destination_folder = '../training_val'  # Replace with the folder to store matching pictures
copy_matching_pictures(trainval_txt, source_folder, destination_folder)

destination_folder = '../test'  # Replace with the folder to store matching pictures
copy_matching_pictures(test_txt, source_folder, destination_folder)
```
